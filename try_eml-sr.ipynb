```python
from eml_sr import Searcher

# Your data
X = [[1.0], [2.0], [3.0]]
y = [2.5, 4.5, 6.5]  # f(x) = 2x + 0.5

# Search for the formula
searcher = Searcher()
result = searcher.fit(X, y)

print(f"Formula: {result.formula}")
result.to_latex()  # Output: Formula: (v_{0} * 2.0) + 0.5
```

```python
from eml_sr import Searcher

X = [[-3],[-2],[-1],[0],[1],[2],[3]]
y = [0.1, 0.2, 0.5, 1, 0.5, 0.2, 0.1]  # f(x) = 1/(1+x^2)

searcher = Searcher()
result = searcher.fit(X, y)

result.to_latex()  # Output: Formula: 1 / (v_{0}^2 + 1)
```
